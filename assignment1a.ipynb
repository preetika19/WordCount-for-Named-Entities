```
import re
import nltk
from nltk.corpus import stopwords
from nltk.tag import pos_tag
```

```
nltk.download('stopwords')
```

```
book_words = book.flatMap(lambda x: re.split("!\"#\＄%&\'\(\)\*\+,-\./:;<=>\?@\[\\\]\^_`{\|}~", x.lower())).filter(lambda x: len(x)>0)
book_words.collect()
```

```
book_stopwords = book_words.filter(lambda x: x not in nltk.corpus.stopwords.words("english"))
book_stopwords.collect()
```

```
named_entities = book_stopwords.map(lambda x: (x,1))
named_entities.collect()
```

```
word_count = named_entities.reduceByKey(lambda x,y: x + y).sortBy(lambda x: -x[-1])
word_count.collect()
```
